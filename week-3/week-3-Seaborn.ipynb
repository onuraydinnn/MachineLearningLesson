```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
example_data = sns.load_dataset("flights")
example_data = example_data.pivot("month","year","passengers")

#Heatmap
sns.heatmap(data=example_data, annot=True, fmt="d", cmap="OrRd")
plt.show()
```

```python
example_data = sns.load_dataset("flights")

#Line Plot
sns.lineplot(data=example_data, x="month", y="passengers")
plt.show()
```

```python
example_data = sns.load_dataset("iris")

#Swarm Plot
sns.swarmplot(data=example_data, x="species", y="petal_length")
plt.show()
```

```python
example_data = sns.load_dataset("titanic")

#Bar Plot
sns.countplot(data=example_data, x="class")
plt.show()
```

```python
example_data = sns.load_dataset("tips")

#Scatter Plot
sns.scatterplot(data=example_data, x="total_bill", y="tip")
plt.show()
```

```python
example_data = sns.load_dataset("titanic")

#Facet Plot
x = sns.FacetGrid(example_data, col="sex")
x.map(sns.histplot, "age")
plt.show()
```

```python
example_iris = sns.load_dataset("iris")

x = sns.clustermap(example_iris.drop("species", axis=1),cmap="magma", standard_scale=1)
plt.show()
```
