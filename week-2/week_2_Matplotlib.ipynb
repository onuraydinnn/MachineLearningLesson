```python
import matplotlib.pyplot as plt
import numpy as np

#Data
x = np.arange(0,10,0.1)
y = np.sin(x)

#Graph Creation
plt.plot(x,y)

#Determine chart properties
plt.title("Sinüs Grafiği")
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")

plt.show()

```

```python
#Data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = np.random.randint(50, 1500, 50)

plt.scatter(x, y, c=colors, s=sizes)

plt.title("Nokta Grafiği")
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")

plt.show()
```

```python
#Data
data = np.random.randn(1000)

plt.hist(data, bins=30)

plt.title("Histogram")
plt.xlabel("Değerler")
plt.ylabel("Frequency")

plt.show()
```

```python
#Data
x = ['A','B','C','D','E']
y = np.random.randint(1,10,5)

plt.bar(x,y)

plt.title("Sinüs Grafiği")
plt.xlabel("Kategoriler")
plt.ylabel("Değerler")

plt.show()
```
